```
import matplotlib as plt
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
from PIL import Image
from six.moves import urllib
plt.style.use("dark_background")
url='https://www.caracteristicas.co/wp-content/uploads/2018/09/lobo-1-e1580435761506.jpg'
imagen_lobo=Image.open(urllib.request.urlopen(url))
```

```
#Transformar la imagen en un array separado
imagen1=np.array(imagen_lobo)
imagen2=np.array(imagen_lobo)
imagen3=np.array(imagen_lobo)
```

```
#Mostrar imagen
plt.figure(figsize=(10,20))
plt.imshow(imagen1)
```

```
#Crear mascara
print(imagen1.shape)
total_rows, total_cols, total_layers=imagen1.shape
X, Y=np.ogrid[:total_rows,:total_cols]
print("X= ",X.shape, " Y= ", Y.shape)
```

```
center_rows, center_cols= total_rows/2,total_cols/2
print(center_rows,center_cols)
```

```
#Distancia al centro
center_distan= (X-center_rows)**2+(Y-center_cols)**2
print(center_distan)
print(np.amax(center_distan))
```

```
#Tama√±o del radio
radius=(total_rows/2.5)**2
print("\n El radio es, ", radius)
```

```
#Crear el filtro circular
circular_mask=(center_distan>radius)
print(circular_mask)
```

```
#aplicamos el filtro
imagen1[circular_mask]=0
plt.figure(figsize=(10,20))
plt.imshow(imagen1)
```

Filtros de Color


```
url2='https://s.libertaddigital.com/2018/10/28/954/0/lobo-animales-salvajes-mundo-animal.jpg'
image_lobo2=Image.open(urllib.request.urlopen(url2))
imagelobo1=np.array(image_lobo2)
plt.figure(figsize=(10,20))
plt.imshow(imagelobo1)
```

```
#Creaci√≥n mascara de colores
green_mask=imagelobo1[:,:,1]>115

#aplicamos mascara

imagelobo1[green_mask,0]=0
imagelobo1[green_mask,1]=0
imagelobo1[green_mask,2]=0
plt.figure(figsize=(10,20))
plt.imshow(imagelobo1)
```
